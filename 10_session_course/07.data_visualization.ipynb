seaborn acepta 2 formatos:  
 1. `sns.algunplot(x=dataset.column1)`
 2. `sns.algunplot(x="column1", data=dataset)`
  

Aquí usamos principalmente la opción `1`

```python
# equiv
sns.countplot(x="island", data=penguins, palette=["lightblue", (212 / 255, 13 / 255, 68 / 255), "#65FA8C"])
```

`palette` establece la paleta de colores

## Box plot

Útil para una variable numérica continua

Menos granular que un histograma

Marca los cuartiles

```python
sns.boxplot(x="Age", data=titanic)
```

Las líneas qué marcan?:  
 - La central: la mediana
 - Borde izquierdo: $Q_1$: percentil 25%
 - Borde derecho: $Q_3$: percentil 75%
 - Izquierda: $Q_1 - 1.5IQR$
 - Derecha: $Q_3 + 1.5IQR$
 
donde $IQR=Q_3-Q_1$

```python
titanic.Age.dropna().quantile([0, 0.25, 0.5, 0.75, 1])
```

Aquí, para agrupar por una variable categórica, usamos `x` (y no `hue`)

```python
sns.boxplot(x="Pclass", y="Age", data=titanic)
```

## Scatterplot

Útil para dos variables numéricas continuas

```python
penguins.head()
```

```python
penguins.corr()
```

```python
sns.scatterplot(x="bill_length_mm", y="flipper_length_mm", data=penguins)
```

```python
sns.scatterplot(x="body_mass_g", y="flipper_length_mm", data=penguins)
```

```python
sns.scatterplot(x="body_mass_g", y="flipper_length_mm", hue="species", data=penguins)
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=penguins)
```

Cuidado con correlaciones agregadas! [Paradoja de Simpson](https://es.wikipedia.org/wiki/Paradoja_de_Simpson#Discriminaci%C3%B3n_por_g%C3%A9nero_en_Berkeley)

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", hue="species", data=penguins)
```

```python
penguins.corr()
```

```python
for s in penguins.species.unique():
    corr = penguins[penguins.species == s].corr()["bill_length_mm"]["bill_depth_mm"].round(2)
    print(f"En la especie {s} la corr entre largo y profundo es {corr}")
```

Podemos usar el argumento `style` para segmentar por una variable categórica (al igual que hacemos con `hue` para el color)

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", style="island", data=penguins)
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", hue="species", style="island", data=penguins)
```

**NOTA**: ciudado con meter mucha información en un gráfico y hacerlo poco comprensible

`size` también puede utilizarse

```python
# size dependent on numerical feature
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", hue="species", size="body_mass_g", data=penguins)
```

## Jointplot

jointplot = scatterplot + histogramas unidimensionales

```python
sns.jointplot(x="bill_length_mm", y="bill_depth_mm", data=penguins)
# sns.jointplot(x=penguins.bill_length_mm, y=penguins.bill_depth_mm)
```

Quieres `hue`? Pues tómalo!

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
```
